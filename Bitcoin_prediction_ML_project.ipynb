```python
#import the libraries
import pandas as pd
import numpy as np
```

```python
#store the data into a variable
df = pd.read_csv('BTC-USD.csv')
#show the first 7 rows of data
df.head(7)
```

```python
df.drop(['Date'], axis=1, inplace=True)
```

```python
df.head(7)
```

```python
# A variable for predicting 'n' days out in the future
prediction_days = 30
# create another column shifted 'n' units up
df['Prediction'] = df[['Adj Close']].shift(-prediction_days)
```

```python
df.head(7)
```

```python
df.tail(7)
```

```python
#Create the independent data set
#conveet the dataFrame to a numpy array and drop the prediction column
X = np.array(df.drop(['Prediction'], axis=1))
#Remove the last 'n' rows where 'n' is the prediction days
X =X[:len(df)-prediction_days]
print(X)
```

```python
#Create the dependent data set
#Convert the dataframe to a numpy array
y = np.array(df['Prediction'])
#Get all of the values except the last 'n' rows
y = y[:-prediction_days]
print(y)
```

```python
#Split the data into 80% training and 20% testting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
#set the prediction_days_array equal to the last 30 rows from the orginal data set
prediction_days_array = np.array(df.drop(['Prediction'], axis=1))[-prediction_days:]
print(prediction_days_array)
```

```python
from sklearn.svm import SVR
#Create and train the support vector machine (Regression) using radial basis function
svr_rbf =SVR(kernel='rbf', C=1e3, gamma=0.00001)
svr_rbf.fit(x_train, y_train)
```

```python
#Test the model
svr_rbf_confidence = svr_rbf.score(x_test, y_test)
print("svr_rbf accuracy: ", svr_rbf_confidence)
```

```python
#Print the predicte values
svm_prediction = svr_rbf.predict(x_test)
print(svm_prediction)
print()

#Print the actual values
print(y_test)
```

```python
#Print the model predictions for the next 'n=30' days
svm_prediction = svr_rbf.predict(prediction_days_array)
print(svm_prediction)
print()

#Print the actual price for Bitcoin for the last 30 days
print(df.tail(30))
```
